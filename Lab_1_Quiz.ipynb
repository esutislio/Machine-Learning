```python
# Create a figure with 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(15,6))

# Plot Price vs. Date
ax1.plot(df.date, df.price, 'mo-', markersize =2)
ax1.set_ylabel('Price')
ax1.set_title('Price vs. Date')

# Plot Cost vs. Date
ax2.plot(df.date, df.cost,  'yo-', markersize =2)
ax2.set_ylabel('Cost')
ax2.set_title('Cost vs. Date')

# Plot Quantity vs. Date
ax3.plot(df.date, df.quantity,  'co-', markersize =2)
ax3.set_ylabel('Quantity')
ax3.set_xlabel('Date')
ax3.set_title('Quantity vs. Date')

# Rotate x-axis labels
ax1.tick_params(axis='x', rotation=45)
ax2.tick_params(axis='x', rotation=45)
ax3.tick_params(axis='x', rotation=45)

# Improve spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
